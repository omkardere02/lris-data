```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine the optimal number of clusters (k) using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (within-cluster sum of squares)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
# Based on the Elbow curve, let's choose k=3 (as it's the "elbow" point)

# Apply K-means clustering with k=3
kmeans_final = KMeans(n_clusters=3, random_state=42)
kmeans_final.fit(X_scaled)
```

```python
# Add cluster labels to the original dataset
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['Cluster'] = kmeans_final.labels_
```

```python
# Visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['Cluster'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-means Clustering of Iris Dataset')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# prompt: prediction on this k-means clustering of lris dataset

import numpy as np
# Predict the cluster labels for new data points
new_data = np.array([[5.5, 3.8, 1.7, 0.4], [6.3, 3.3, 6.0, 2.5]])
new_data_scaled = scaler.transform(new_data)
predicted_labels = kmeans_final.predict(new_data_scaled)
print('Predicted labels for new data points:')
print(predicted_labels)
```
